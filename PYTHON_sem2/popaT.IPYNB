```python
import matplotlib.pyplot as plt
x = [3, 1, 3] 
y = [3, 2, 1]
plt.plot(x, y)
plt.title("Line Chart")
plt.legend(["Li"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3] 
y = [2,3,1]
plt.stem(x, y) 
plt.title("Line Chart")
plt.legend(["li"])
```

```python
import matplotlib.pyplot as plt
x = ["1","2","3"] 
y = [2,3,1]
plt.bar(x, y) 
plt.title("Line Chart")
plt.legend(["Li"])
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [1, 2,4,3]
plt.hist(x, bins = [1,2,4,6],edgecolor='w')
plt.title("Histogram")
plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3] 
y = [2,3,1]
plt.scatter(x, y,c=[1,0,2]) 
plt.title("Line Chart")
plt.legend(["Li"])
plt.show()
```

```python
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
Studying = [7, 8, 6, 11, 7]
playing = [8, 5, 7, 8, 13]
gaming=[4,2,5,3,6]
plt.stackplot(days, Studying, playing,gaming,colors =['r', 'b','g'])
plt.xlabel('Days')
plt.ylabel('No of Hours')
plt.title('Representation of Study and Playing wrt to Days')
plt.show()
```
